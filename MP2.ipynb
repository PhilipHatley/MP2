```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_red = pd.read_excel(r'C:\Users\phili\OneDrive\Desktop\4. Semester\BI\Aflevering\MP2\winequality-red.xlsx', header=1)
df_white = pd.read_excel(r'C:\Users\phili\OneDrive\Desktop\4. Semester\BI\Aflevering\MP2\winequality-white.xlsx', header=1)

df_red['Type'] = 'Red'
df_white['Type'] = 'White'

df_combined = pd.concat([df_red, df_white], ignore_index=True)

df_combined.dropna(inplace=True)

print("Summary statistics for combined wine data:\n", df_combined.describe())

plt.hist(df_combined[df_combined['Type'] == 'Red']['alcohol'], bins=20, color='red', alpha=0.5, label='Red Wine', edgecolor='black')
plt.hist(df_combined[df_combined['Type'] == 'White']['alcohol'], bins=20, color='blue', alpha=0.5, label='White Wine', edgecolor='black')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.title('Distribution of Alcohol Content in Combined Wine Data')
plt.legend()
plt.show()

plt.boxplot([df_combined[df_combined['Type'] == 'Red']['pH'], df_combined[df_combined['Type'] == 'White']['pH']], labels=['Red Wine', 'White Wine'])
plt.xlabel('Wine Type')
plt.ylabel('pH')
plt.title('Boxplot of pH for Combined Wine Data')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x='quality', hue='Type', data=df_combined, palette={'Red':'red', 'White':'blue'})
plt.xlabel('Quality Rating')
plt.ylabel('Count')
plt.title('Distribution of Quality Ratings for Combined Wine Data')
plt.legend(title='Wine Type')
plt.show()

plt.scatter(df_combined[df_combined['Type'] == 'Red']['volatile acidity'], df_combined[df_combined['Type'] == 'Red']['fixed acidity'], color='red', alpha=0.5, label='Red Wine')
plt.scatter(df_combined[df_combined['Type'] == 'White']['volatile acidity'], df_combined[df_combined['Type'] == 'White']['fixed acidity'], color='blue', alpha=0.5, label='White Wine')
plt.xlabel('Volatile Acidity')
plt.ylabel('Fixed Acidity')
plt.title('Scatterplot of Volatile Acidity vs. Fixed Acidity for Combined Wine Data')
plt.legend()
plt.show()

numeric_columns = df_combined.select_dtypes(include=['float64', 'int64']).columns
corr_combined = df_combined[numeric_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_combined, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap for Combined Wine Data')
plt.show()
```

```python
sns.boxplot(x='residual sugar', data=df_combined)
plt.title('Boxplot of Residual Sugar')
plt.show()

from scipy import stats
z_scores = stats.zscore(df_combined['residual sugar'])
outliers = df_combined[abs(z_scores) > 3]

print("Rows with outliers in 'residual sugar':")
print(outliers)

df_combined = df_combined[abs(z_scores) <= 3]
```

```python
correlation_with_quality = corr_combined['quality'].drop('quality') 
lowest_correlation_attribute = correlation_with_quality.idxmin()

df_combined.drop(lowest_correlation_attribute, axis=1, inplace=True)
print(f"The attribute with the lowest correlation to wine quality is '{lowest_correlation_attribute}' and it has been removed from the dataset.")
```

```python
df_combined = pd.get_dummies(df_combined, columns=['Type'], drop_first=True)
print("Categorical data transformed into numeric.")
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np 

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_combined.drop('quality', axis=1))

pca = PCA()
pca.fit(scaled_data)

explained_variance_ratio = pca.explained_variance_ratio_

plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio by Number of Components')
plt.show()

cumulative_variance_ratio = np.cumsum(explained_variance_ratio)
optimal_num_components = np.argmax(cumulative_variance_ratio >= 0.95) + 1
print(f"The optimal number of components to retain at least 95% of the variance is {optimal_num_components}.")
```
